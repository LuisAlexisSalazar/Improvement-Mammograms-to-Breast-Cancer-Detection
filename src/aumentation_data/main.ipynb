```python

data_dir_list = os.listdir(path_to_read)
num_images = 0
for name_image in data_dir_list:
    path_image = path_to_read + "/" + name_image
    img = load_img(path_image)
    # plt.imshow(img, cmap='gray')
    # break
    x = img_to_array(img)  # creating a Numpy array with shape (3, 150, 150)
    x = x.reshape((1,) + x.shape)  # converting to a Numpy array with shape (1, 3, 150, 150)
    i = 0

    for batch in datagen.flow(x, batch_size=1, save_to_dir=path_to_save, save_prefix=name_image[:-3],
                              save_format='tif'):
        i += 1
        if i > images_increased:
            num_images += images_increased
            break

print("Total de imagenes generadas:", num_images)
```

```python
def show_imgs_titles(imgs, titles):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

    ax1.imshow(imgs[0], cmap='gray')
    ax1.set_title(titles[0])
    ax1.axis('off')

    ax2.imshow(imgs[1], cmap='gray')
    ax2.set_title(titles[1])
    ax2.axis('off')

    ax3.imshow(imgs[2], cmap='gray')
    ax3.set_title(titles[2])
    ax3.axis('off')

    plt.show()


def read_img_out(file="out2.tif"):
    img = cv2.imread(file)
    b, g, r = cv2.split(img)
    titles = ["Imagen Procesada", "Brint S", "Brint M"]
    imgs = [b, g, r]

    # for i in imgs:
    #     print(i.shape)
    # cv2.imwrite("original.jpg", b)
    show_imgs_titles(imgs, titles)


name_file = ""
path_file_test = "E:\\U\Improvement-Mammograms-to-Breast-Cancer-Detection\data\data_aug\MIAS\ClassBMN\\benigno\\" + name_file

read_img_out(path_file_test)
```
